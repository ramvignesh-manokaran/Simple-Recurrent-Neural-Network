```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### Sentences
sentences=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sentences
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### One Hot Representation
one_hot_repr = [one_hot(words,voc_size) for words in sentences]
one_hot_repr
```

```python
## Word Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sentences_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sentences_length)
print(embedded_docs)
```

```python
## feature representation
dimension = 10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dimension,input_length=sentences_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
